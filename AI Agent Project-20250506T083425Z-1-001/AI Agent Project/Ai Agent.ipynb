```python
import os

secrets_path = "/content/drive/MyDrive/AI Agent Project/secrets.txt"

with open(secrets_path, "r") as f:
    for line in f:
        if line.strip():
            key, value = line.strip().split("=", 1)
            os.environ[key.strip()] = value.strip()

# âœ… Check if loaded
print("OPENAI_API_KEY:", os.environ.get("OPENAI_API_KEY")[:6] + "..." if os.environ.get("OPENAI_API_KEY") else "Not set")
print("SERPAPI_API_KEY:", os.environ.get("SERPAPI_API_KEY")[:6] + "..." if os.environ.get("SERPAPI_API_KEY") else "Not set")
```

```python
from langchain_openai import ChatOpenAI
from langchain_community.utilities import SerpAPIWrapper
from langchain.agents import Tool, initialize_agent, AgentType

# Initialize LLM
llm = ChatOpenAI(
    model="gpt-4",
    temperature=0.7,
    api_key=os.environ["OPENAI_API_KEY"]
)

# âœ… Use SerpAPI for real web search
search = SerpAPIWrapper(serpapi_api_key=os.environ["SERPAPI_API_KEY"])

tools = [
    Tool(
        name="Search",
        func=search.run,
        description="Searches the web for latest news or information"
    )
]

agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)

# Test the agent
query = "Recent breakthroughs in generative AI for education"
response = agent.run(query)
print("\nğŸ” Agent Response:\n", response)
```

```python
from memory_store import get_chroma_memory

# Initialize Chroma vector store
memory_db = get_chroma_memory()

# Add memory: simulate past topics and outputs
texts = [
    "Generative AI can create personalized lesson plans in education.",
    "In healthcare, AI is being used to detect diseases earlier.",
    "Cybersecurity uses AI for threat detection and anomaly analysis."
]
metadatas = [
    {"topic": "education"},
    {"topic": "healthcare"},
    {"topic": "cybersecurity"}
]

memory_db.add_texts(texts, metadatas=metadatas)
```

```python
results = memory_db.similarity_search("How is AI used in schools?", k=2)

for i, r in enumerate(results, 1):
    print(f"\nğŸ” Match #{i}")
    print("Text:", r.page_content)
    print("Metadata:", r.metadata)
```

```python
from langchain.tools import BaseTool

class MemorySearchTool(BaseTool):
    name: str = "MemorySearch"
    description: str = "Searches past stored knowledge from memory."

    def _run(self, query: str) -> str:
        results = memory_db.similarity_search(query, k=2)
        if not results:
            return "No relevant memory found."
        return "\n".join([f"- {r.page_content}" for r in results])

    def _arun(self, query: str):
        raise NotImplementedError("Async not supported.")
```

```python
tool = MemorySearchTool()
print(tool._run("What have you stored about AI in education?"))
```
